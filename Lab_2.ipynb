# Exercise 1) Write Python functions for the three metrics (Euclidean, Manhattan and Minkowski):

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

def euclidean(a, b):
    return np.sqrt(np.sum((a - b) ** 2))

def manhattan(a, b):
    return np.sum(np.abs(a - b))

def minkowski(a, b, p=3):
    return np.sum(np.abs(a - b) ** p) ** (1/p)
```

```python
a = np.array([1, 2])
b = np.array([4, 6])
```

* Minkokski (p = 3) ~ 4.4979

Looking at the plots, there are clear differences in how these two distance metrics perform:

**Euclidean distance** works much better for this clustered data. It correctly identifies the two artificially injected outliers that are spatially far from the main clusters. This makes sense because Euclidean distance measures the actual geometric separation between points, which is exactly what we want when detecting spatial outliers.

**Cosine distance** performs poorly in this scenario. It either fails to identify the true outliers or incorrectly flags points within the clusters as outliers. This happens because cosine distance measures the angle between vectors from the origin, not their actual spatial separation. For points that are clustered based on their position (not their direction from origin), this angular measurement is not very meaningful.

The cosine distance has a harder time with outliers in this case because it's designed for high-dimensional data where direction matters more than magnitude, such as text or image embeddings. For geometric clustering problems like this one, Euclidean distance is the more appropriate choice.

```python
from sklearn.metrics import pairwise_distances

# --- 1) Make clustered data ---
X, y = make_blobs(n_samples=300,
                  centers=[(-2, -2), (0, 4)],
                  cluster_std=0.6,
                  random_state=42)

# --- 2) Inject outliers ---
n_outliers = 2
rng = np.random.RandomState(42)

# choose a wide box far outside clusters
X_outliers = rng.uniform(low=-12, high=10, size=(n_outliers, 2))


# combine
X_full = np.vstack([X, X_outliers])
y_full = np.hstack([y, [-1]*n_outliers])  # mark outliers as -1

# --- Compute pairwise Euclidean distances ---
D = pairwise_distances(X_full, metric='euclidean')

# Distance profile: average distance to all other points
mean_dist = D.mean(axis=1)

# Identify potential outliers (top-2 by average distance)
outlier_indices = np.argsort(mean_dist)[-2:][::-1]

# --- 3) Plot ---
plt.scatter(X_full[:, 0], X_full[:, 1], c=y_full, cmap="tab10", s=40)
plt.scatter(X_full[outlier_indices,0], X_full[outlier_indices,1],
            facecolors="none", edgecolors="red",
            s=120, linewidths=2, label="outliers")

plt.legend()
plt.title("make_blobs clusters + clear outliers")
plt.show()
```

**Analysis for linear data:**

This exercise shows a dramatic failure of cosine distance for linear data. The Euclidean distance correctly identifies the two true outliers that were placed far from the line. However, the cosine distance completely misses the actual outliers and instead identifies points at the extremes of the line itself as outliers.

This happens because points on the line y = 2x + 1 have very different angles when measured from the origin (0,0). Points in the negative quadrant have angles around 180 degrees, while points in the positive quadrant have angles around 0 degrees. To the cosine distance metric, these extreme points on the line appear most "different" from the average direction of all points, even though they're actually part of the same linear pattern.

This is a perfect example of when not to use cosine distance - it's completely inappropriate for geometric data where spatial relationships matter more than angular relationships from the origin.

#Exercise 5) In this first cell we will be calculating the cosine and euclidean distances for some simulated image data, we will see that cosine distance is more robust to variations in the data

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_distances, euclidean_distances

rng = np.random.default_rng(0)

# Simulate 512-d image embeddings
# Image A: a "reference image"
img_A = rng.normal(size=(1, 512))

# Image B: same class (embedding points in nearly the same direction, scaled)
img_B = img_A * 5.0 + rng.normal(scale=0.01, size=(1, 512))

# Image C: different class (independent random vector)
img_C = rng.normal(size=(1, 512))

X = np.vstack([img_A, img_B, img_C])

# Compute cosine vs Euclidean distances
cos_dist = cosine_distances(X)
euc_dist = euclidean_distances(X)

print("Cosine distance matrix:\n", np.round(cos_dist, 3))
print("\nEuclidean distance matrix:\n", np.round(euc_dist, 3))
```

```python
# Use only two features: mean radius and mean area
X2 = X[['area error', 'mean area']].values

# --- Compute pairwise Euclidean distances ---
D = pairwise_distances(X2, metric='euclidean')

# Distance profile: average distance to all other points
mean_dist = D.mean(axis=1)

# Identify potential outliers (top-5 by average distance)
outlier_indices = np.argsort(mean_dist)[-5:][::-1]
print("Top 5 outliers by mean distance:", outlier_indices)

# --- Plot ---
plt.figure(figsize=(8,6))
plt.scatter(X2[:,0], X2[:,1], c="lightblue", s=40, edgecolor="k", alpha=0.7)
plt.scatter(X2[outlier_indices,0], X2[outlier_indices,1],
            c="red", marker="*", s=200, label="Outliers")
plt.xlabel("Area error")
plt.ylabel("Mean Area")
plt.title("Breast Cancer Dataset: Outlier Detection via Euclidean Distance")
plt.legend()
plt.show()
```

**Analysis of breast cancer dataset outliers:**

The Euclidean and cosine distances identify different sets of outliers in this real dataset, which is actually quite interesting. The Euclidean distance focuses on points that are far away in terms of absolute values - these are samples with unusually high area error or mean area measurements. These represent cases that are genuinely extreme in magnitude.

The cosine distance identifies points that deviate from the proportional relationship between the two variables. These might be cases where the ratio between area error and mean area is unusual, even if the absolute values aren't extremely high.

Both types of outliers are valid depending on what we're looking for. Medical outliers based on absolute measurements (Euclidean) might indicate measurement errors or truly exceptional cases, while proportional outliers (cosine) might indicate different underlying biological patterns.


```python
# --- Compute pairwise Cosine distances ---
D = pairwise_distances(X2, metric='cosine')

# Distance profile: average distance to all other points
mean_dist = D.mean(axis=1)

# Identify potential outliers (top-5 by average distance)
outlier_indices = np.argsort(mean_dist)[-5:][::-1]
print("Top 5 outliers by mean distance:", outlier_indices)

# --- Plot ---
plt.figure(figsize=(8,6))
plt.scatter(X2[:,0], X2[:,1], c="lightblue", s=40, edgecolor="k", alpha=0.7)
plt.scatter(X2[outlier_indices,0], X2[outlier_indices,1],
            c="red", marker="*", s=200, label="Outliers")
plt.xlabel("Area error")
plt.ylabel("Mean Area")
plt.title("Breast Cancer Dataset: Outlier Detection via Cosine Distance")
plt.legend()
plt.show()
```

### What are some of your observation about these heatmaps? Are some plots similar to others, additionally, does any of the heatmaps appear to struggle with detecting any similairty between the distances of a large amount of points?

**Observations about the heatmaps:**

The **Euclidean and Manhattan distance heatmaps look very similar** to each other. This makes sense because both measure actual distances in the feature space, just calculated differently (Manhattan uses L1 norm while Euclidean uses L2 norm). They show gradual color transitions indicating how different samples are from each other spatially.

The **cosine distance heatmap is dramatically different** from the other two. Most of the heatmap appears very dark (low distances), meaning most points have similar angles from the origin. Only a few specific samples stand out as having very different directions in the feature space.

**The cosine distance heatmap appears to struggle with detecting similarity** between a large amount of points because it compresses most relationships into a narrow range of very small distances. This makes it difficult to distinguish between most samples, as they appear almost identical in terms of cosine distance. This confirms that cosine distance is not well-suited for this type of low-dimensional geometric data where the magnitude and position of features matter more than their directional relationships from the origin.


```python
import seaborn as sns

# Select just two features
X2 = X[['area error', 'mean area']].values
X2 = X2[:50,:]

# --- Compute pairwise distances ---
D_euclidean = pairwise_distances(X2, metric='euclidean')
D_manhattan = pairwise_distances(X2, metric='manhattan')
D_cosine    = pairwise_distances(X2, metric='cosine')

print("Distance matrix shape:", D_euclidean.shape)  # (569, 569)

# --- Heatmap plotting function ---
def plot_heatmap(D, title):
    plt.figure(figsize=(8,6))
    sns.heatmap(D, cmap="viridis", cbar_kws={'label': 'Distance'})
    plt.title(title)
    plt.xlabel("Sample index")
    plt.ylabel("Sample index")
    plt.show()

# Plot heatmaps for each metric
plot_heatmap(D_euclidean, "Pairwise Euclidean Distance Heatmap")
plot_heatmap(D_manhattan, "Pairwise Manhattan Distance Heatmap")
plot_heatmap(D_cosine,    "Pairwise Cosine Distance Heatmap")
```
